## Aufgabe 01: Endgeschwindigkeit beim freien Fall

### Teil 1: Berechnung mit for-Schleife

```python
import math

g = 9.81
h = 10

for h in range(1,11):
    v = math.sqrt(2 * g * h)
    print("h = ", h , " m; v = " , v , " m/s")
```

#### Einbindung von `numpy` f√ºr arrays

```python
import numpy as np

x = np.arange(1,11)
print(x)

y = x**2
print(y)
```

### Teil 2: Diagramm
#### Graphik anzeigen

```python
from matplotlib import pyplot as plt

plt.plot(x,y, color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.title("y=x^2")

plt.show()
```

## Aufgabe 02: Fallender Ball- Geschwindigkeit und H√∂he berechnen

```python
import math
from matplotlib import pyplot as plt
import numpy as np

g = 9.81
h = 100
t = 0

while h > 0:
    t += 0.25
    h = 100 - 0.5 * g * t**2
    print("h: ", h , "m \t t: ", t, "s")
    
t = np.arange(0, t, 0.1)
h = 100 - 0.5 * g * t**2

plt.plot(t,h, color="red")
plt.xlabel("t")
plt.ylabel("h")
plt.grid()
plt.title("Freier Fall aus 100m")

plt.show()
```
