```python
import math
import numpy as np
import matplotlib.pyplot as plt

g = 9.81

m1 = 0  # Masse
l1 = 0.20  # Länge
w1 = math.sqrt(g/l1)  # Auslenkung vom Lot

x0 = 0.5  # Startauslenkung
x1 = 0  # Ort
t1 = 0  # Zeit

tges = []
xges = []



for i in range(0,1000):
    x1 = x0 * math.cos(math.radians(w1*t1))
    tges.append(t1)
    xges.append(x1)
    t1 = t1 + 0.1

plt.plot(tges,xges, color='blue')
plt.grid()
plt.xlabel("t in s")
plt.ylabel("Höhe in m")
plt.title("Fadenpendel")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import HTML, display
import matplotlib.animation as animation
import math

fig, ax = plt.subplots()
t = np.linspace(0, 3, 40)
g = -9.81  # acceleration (negative for downward)

l1 = 0.20  # Länge
# use absolute value of g so the argument of sqrt is non-negative
w1 = math.sqrt(abs(g) / l1)  # Eigenfrequenz (rad/s)

x0 = 0.5
# compute displacement as an array over t (use numpy cos, no degree conversion)
x1 = x0 * np.cos(w1 * t)

scat = ax.scatter(t[0], x1[0], c="b", s=5, label=f'x0 = {x0} m')
line2 = ax.plot(t[0], x1[0], label=f'x0 = {x0} m')[0]
ax.set(xlim=[0, 3], ylim=[-0.6, 0.6], xlabel='Time [s]', ylabel='Displacement [m]')
ax.legend()


def update(frame):
    # for each frame, update the data stored on each artist.
    x = t[:frame]
    y = x1[:frame]
    # update the scatter plot:
    data = np.stack([x, y]).T
    scat.set_offsets(data)
    # update the line plot:
    line2.set_xdata(t[:frame])
    line2.set_ydata(x1[:frame])
    return (scat, line2)


ani = animation.FuncAnimation(fig=fig, func=update, frames=len(t), interval=30)

try:
    display(HTML(ani.to_jshtml()))
except Exception as e:
    display(HTML(ani.to_html5_video()))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import HTML, display
import matplotlib.animation as animation

fig, ax = plt.subplots()
t = np.linspace(0, 3, 40)
g = -9.81
v0 = 12
z = g * t**2 / 2 + v0 * t

v02 = 5
z2 = g * t**2 / 2 + v02 * t

scat = ax.scatter(t[0], z[0], c="b", s=5, label=f'v0 = {v0} m/s')
line2 = ax.plot(t[0], z2[0], label=f'v0 = {v02} m/s')[0]
ax.set(xlim=[0, 3], ylim=[-4, 10], xlabel='Time [s]', ylabel='Z [m]')
ax.legend()


def update(frame):
    # for each frame, update the data stored on each artist.
    x = t[:frame]
    y = z[:frame]
    # update the scatter plot:
    data = np.stack([x, y]).T
    scat.set_offsets(data)
    # update the line plot:
    line2.set_xdata(t[:frame])
    line2.set_ydata(z2[:frame])
    return (scat, line2)


ani = animation.FuncAnimation(fig=fig, func=update, frames=40, interval=30)

try:
    display(HTML(ani.to_jshtml()))
except Exception as e:
    display(HTML(ani.to_html5_video()))

#plt.show()

```
