# Freier Fall
### Teil 1: Ohne Luftwiderstand

```python
import math
import numpy as np
import matplotlib.pyplot as plt

v = 0
t = 0
s = 0
sg = 0
dt = 0.01
tr = []
sr = []
vr = []
sr1 = []

for i in range(0,100):
    t = t + dt 
    a = 9.81
    v = v+a*dt
    sg = 1/2*a*t**2 
    s = s+v*dt
    tr.append(t)
    vr.append(v)
    sr.append(s)
    sr1.append(sg)


    
#tr = np.array(tr)
#sr = np.array(sr)

plt.plot(tr,sr, color='red')
plt.plot(tr,sr1, color='blue')
plt.grid()
plt.xlabel("t in s")
plt.ylabel("verstrichene Strecke in m")
plt.title("Freier Fall ohne Luftwiderstand")
plt.show()
```

### Teil 2: Mit Luftwiderstand

```python
import math
import numpy as np
import matplotlib.pyplot as plt

v = 0
t = 0
h = 0
a = 0
dt = 0.01

vr = []
tr = []
hr = []

m = 80
A = 1
g = 9.81

p = 1.164   #Luft dichte
c = 0.78     #CW Wert

def Fd(v):
    fd = 1/2*p+v**2*c*A #Luftwiderstand
    if v<0:
        fd = -fd
    return fd


#Fg = m*g                #if Fg-Fd == 0 -> v == konstant

for i in range(0,1000):
    t = t + dt 
    v = v+a*dt

    a = g-Fd(v)/m
    h = h+v*dt

    vr.append(v)
    tr.append(t)
    hr.append(h)

plt.plot(tr, hr, color='red')
plt.xlabel("t in s")
plt.ylabel("h in m")
plt.grid()
plt.title("Freier Fall mit Luftwiderstand")
plt.show()

```
