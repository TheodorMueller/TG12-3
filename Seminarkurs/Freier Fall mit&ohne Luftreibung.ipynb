# Freier Fall
### Ohne Luftwiderstand

```python
import math
import numpy as np
import matplotlib.pyplot as plt

v = 0
t = 0
s = 0
sg = 0
dt = 0.01
tr = []
sr = []
vr = []
sr1 = []

for i in range(0,100):
    t = t + dt 
    a = 9.81
    v = v+a*dt
    sg = 1/2*a*t**2 
    s = s+v*dt
    tr.append(t)
    vr.append(v)
    sr.append(s)
    sr1.append(sg)


    
#tr = np.array(tr)
#sr = np.array(sr)

plt.plot(tr,sr, color='red')
plt.plot(tr,sr1, color='blue')
plt.grid()
plt.xlabel("t in s")
plt.ylabel("verstrichene Strecke in m")
plt.title("Freier Fall ohne Luftwiderstand")
plt.show()
```

### Mit Luftwiderstand

```python
import math
import numpy as np
import matplotlib.pyplot as plt

vL = 0
tL = 0
sL = 0
dt = 0.01
tLr = []
sLr = []
vLr = []

p = 1.164   #Luft dichte
c = 1.3     #CW Wert
A = 25      # m^2  Querschnitt

a = 9.81
g = 9.81
m = 90

#Fd = 1/2*p+v**2*c*A #Luftwiderstand
Fg = m*g                #if Fg-Fd == 0 -> v == konstant

for i in range(0,100):
    tL = tL + dt 
    vL = vL+a*dt
    if a>0:
        sL = 1/2*a*tL**2 
    else:
        sL = vL*tL

    Fd = 1/2*p+vL**2*c*A
    Fg = m*g
    if Fg - Fd <= 0:
        a = 0
    
    tLr.append(tL)
    vLr.append(vL)
    sLr.append(sL)


    
#tr = np.array(tr)
#sr = np.array(sr)

plt.plot(tLr,sLr, color='red')
plt.grid()
plt.xlabel("t in s")
plt.ylabel("verstrichene Strecke in m")
plt.title("Freier Fall mit Luftwiderstand")
plt.show()
```
