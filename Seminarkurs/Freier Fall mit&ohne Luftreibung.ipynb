# Freier Fall
### Ohne Luftwiderstand

```python
import math
import numpy as np
import matplotlib.pyplot as plt

v = 0
t = 0
s = 0
sg = 0
dt = 0.01
tr = []
sr = []
vr = []
sr1 = []

for i in range(0,100):
    t = t + dt 
    a = 9.81
    v = v+a*dt
    sg = 1/2*a*t**2 
    s = s+v*dt
    tr.append(t)
    vr.append(v)
    sr.append(s)
    sr1.append(sg)


    
#tr = np.array(tr)
#sr = np.array(sr)

plt.plot(tr,sr, color='red')
plt.plot(tr,sr1, color='blue')
plt.grid()
plt.xlabel("t in s")
plt.ylabel("verstrichene Strecke in m")
plt.title("Freier Fall ohne Luftwiderstand")
plt.show()
```
