# Lektion 3: Bewegung in 2 Dimensionen
## Schiefer Wurf
### Teil 1


```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Schräger Wurf ohne Luftwiderstand
g = 9.81
t = 0
dt = 0.001   # delta t
v0 = 15
w = math.radians(60)  # radian vom winkel

x = 0
y = 0
xi = 0
yi = 0

xR = [0]
yR = [0]

xRi = [0]
yRi = [0]

vx = math.cos(w)*v0
vy = math.sin(w)*v0
vi = vy
print(vx)

for i in range(0,6000):  #310 für 60°
    t = t + dt
    x = vx*t
    y = vy*t-1/2*g*t**2

    # Kleine Schritte
    vi = vi+(-g)*dt
    yi = yi+vi*dt

    xi = xi+vx*dt

    xR.append(x)
    yR.append(y)
    xRi.append(xi)
    yRi.append(yi)
    try:
        if y < 0:
            print(math.sqrt(vx**2+vy**2))
            break
    except:
        pass

plt.plot(xR, yR)
plt.plot(xRi, yRi, color="red")
plt.title("Schiefer Wurf")
plt.grid()
plt.xlabel("x in m")
plt.ylabel("y in m")
plt.ylim(0)
plt.show()
```
