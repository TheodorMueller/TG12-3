```python
class Lagerverwaltung:
    def __init__(self):
        aAnz:int = 0
        

    def __findeLagerplatz(self,pArtikelname:str) -> int:
        idx:int = 0
        name:str = ""
        frei:bool 
        idxGefundenerLp:int = -1
        gefunden:bool = False

        for idx in range(0,self.aAnz-1):
            name = self.derLagerplatz[idx].gibArtikelname()
            if name == pArtikelname:
                gefunden = True
                idxGefundenerLp = idx
                self.dieGui.melden("Lagerplatz ", idxGefundenerLp, " gefunden")

        if gefunden == False:
            for idx in range(0,self.aAnz-1):
                frei = self.derLagerplatz[idx].istFrei()
                if frei == True:
                    gefunden = True
                    idxGefundenerLp = idx
                    self.dieGui.melden("Ware wird eingelagert")

        if gefunden == False:
            self.dieGui.melden("Lager voll, einlagern nicht m√∂glich")
        return idxGefundenerLp
    
```
