## Server aufsetzen:
---
1.  ```Python
    app = Flask(__name__)
    ```

2. Routen vom Server für verschiedene Funktionen:

    ```Python
    @app.route('/abc', methods=['POST']) # Oder 'GET'   | Wenn man methods weg lässt ist es autom. 'GET'
    def abc():
        # das was die Funktion machen soll
        return jsonify({}) # dictionary
    ```

3. Server starten:

    ```Python
    if __name__ == '__main__':
        app.run(host='0.0.0.0', port=12345, debug=True)     # Port kann man frei wählen | host='0.0.0.0' heist Localhost
    
    # ab hier darf nichts mehr stehen!!!
    ```

## Pydantic:
---
- Zum einfachen erstellen von Klassen
- Bedingungen sind einfach zu erstellen

```Python
class Klasse(BaseModel):
    name: str = Field()
    zahl: int = Field()

    model_config = {"validate_assignmet":True}  #=> Bedingungen in Field() werden dauerhaft geprüft
```

-> Field() bedingungen:
- `le` <=
- `lt` <
- `ge` >=
- `gt` >

### 3. Wichtige OOP-Konzepte

```python
class Objekt:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
```

-   **Konstruktor:** `__init__`
-   **Kapselung:** private Attribute `__name`
-   **Getter/Setter:** via `@property`
-   **Vererbung:** `class Subclass(BaseClass)`

## Datenhaltung: Objekte in einer Liste verwalten
```Python
liste = []

liste.append(Objekt(1, 2, 3))

for obj in liste:
    print(obj.a, obj.b, obj.c)
```
