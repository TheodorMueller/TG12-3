# Datenbanksysteme: n:m-Beziehungen in UML, Tabellen und PostgreSQL


## Reales Beispiel: Musikschule

Eine Musikschule verwaltet **Schülerinnen und Schüler**, **Kurse** und **Lehrkräfte**.

- Eine **Lehrkraft** kann mehrere Kurse anbieten.
- Ein **Kurs** wird genau von einer Lehrkraft geleitet.
- Eine **Schülerin bzw. ein Schüler** kann mehrere Kurse besuchen.
- Ein **Kurs** kann von mehreren Schülerinnen und Schülern besucht werden.
- Die Beziehung zwischen `Schueler` und `Kurs` ist also eine **n:m-Beziehung**.
- Für jede Teilnahme soll zusätzlich gespeichert werden, seit wann die Person im Kurs ist und welchen Status die Teilnahme hat.

Wir modellieren also drei Klassen:

1. `Schueler`
2. `Kurs`
3. `Lehrkraft`

## Musterlösung 1: UML-Klassendiagramm

Eine mögliche textuelle UML-Darstellung in Mermaid:

```mermaid
classDiagram
    class Schueler {
        +int schuelerId
        +String vorname
        +String nachname
        +String email
        +Date geburtsdatum
    }

    class Kurs {
        +int kursId
        +String titel
        +String instrument
        +int maxTeilnehmer
        +String wochentag
    }

    class Lehrkraft {
        +int lehrkraftId
        +String vorname
        +String nachname
        +String fachgebiet
    }

    Lehrkraft "1" --> "0..*" Kurs : leitet
    Schueler "0..*" --> "0..*" Kurs : besucht
```

**Begründung:**

- Eine Lehrkraft kann keinen, einen oder mehrere Kurse leiten: `0..*`.
- Jeder Kurs wird genau von einer Lehrkraft geleitet: `1`.
- Ein Schüler kann keinen, einen oder mehrere Kurse besuchen: `0..*`.
- Ein Kurs kann von keinem, einem oder mehreren Schülern besucht werden: `0..*`.
- Die Assoziation `Schueler besucht Kurs` ist damit eine n:m-Beziehung.

## Musterlösung 3: Relationale Tabellenstruktur

Benötigte Tabellen:

1. `lehrkraft`
2. `kurs`
3. `schueler`
4. `teilnahme`

Die n:m-Beziehung zwischen `schueler` und `kurs` wird nicht als einzelne Spalte gespeichert, sondern durch die zusätzliche Tabelle `teilnahme`.

### Relationenschreibweise

- `lehrkraft(`**`lehrkraft_id`**`, vorname, nachname, fachgebiet)`
- `kurs(`**`kurs_id`**`, titel, instrument, max_teilnehmer, wochentag, lehrkraft_id → lehrkraft.lehrkraft_id)`
- `schueler(`**`schueler_id`**`, vorname, nachname, email, geburtsdatum)`
- `teilnahme(`**`schueler_id → schueler.schueler_id`**`, `**`kurs_id → kurs.kurs_id`**`, seit, status)`

Der Primärschlüssel von `teilnahme` ist zusammengesetzt aus `schueler_id` und `kurs_id`. Dadurch kann dieselbe Person nicht zweimal im selben Kurs eingetragen werden.

# Aufgabe 4: Darstellung der n:m-Beziehung erklären

Erklären Sie, warum die n:m-Beziehung zwischen `Schueler` und `Kurs` nicht direkt in einer der beiden Tabellen gespeichert werden sollte.

Beantworten Sie:

1. Was wäre problematisch, wenn man in `schueler` eine Spalte `kurs_ids` speichert?

3. Warum ist eine Zwischentabelle fachlich und technisch besser?

## Musterlösung 4: Auflösung der n:m-Beziehung

Eine n:m-Beziehung wird in einem relationalen Datenbankmodell durch eine **Zwischentabelle** aufgelöst.

Problematische Lösungen wären:

- In `schueler` eine Spalte `kurs_ids` mit mehreren Werten wie `101,103` zu speichern.
- In `kurs` eine Spalte `schueler_ids` mit mehreren Werten wie `201,203,205` zu speichern.

Das verletzt die Idee atomarer Spaltenwerte: In einer Zelle soll genau ein fachlicher Wert stehen. Außerdem wären JOINs, Prüfungen und Änderungen deutlich schwieriger.

Die Zwischentabelle `teilnahme` ist besser, weil:

- jede Teilnahme als eigener Datensatz gespeichert wird,
- `schueler_id` und `kurs_id` als Fremdschlüssel referenzielle Integrität sichern,
- Zusatzinformationen zur Beziehung gespeichert werden können, z. B. `seit` und `status`,
- SQL-Abfragen mit JOIN, GROUP BY und Aggregatfunktionen sauber möglich sind.

## Musterlösung 5: PostgreSQL-DDL

```sql
-- PostgreSQL: Datenbankmodell Musikschule

DROP TABLE IF EXISTS teilnahme;
DROP TABLE IF EXISTS kurs;
DROP TABLE IF EXISTS schueler;
DROP TABLE IF EXISTS lehrkraft;

CREATE TABLE lehrkraft (
    lehrkraft_id   INTEGER PRIMARY KEY,
    vorname        VARCHAR(50) NOT NULL,
    nachname       VARCHAR(50) NOT NULL,
    fachgebiet     VARCHAR(80) NOT NULL
);

CREATE TABLE schueler (
    schueler_id    INTEGER PRIMARY KEY,
    vorname        VARCHAR(50) NOT NULL,
    nachname       VARCHAR(50) NOT NULL,
    email          VARCHAR(120) NOT NULL UNIQUE,
    geburtsdatum   DATE NOT NULL
);

CREATE TABLE kurs (
    kurs_id         INTEGER PRIMARY KEY,
    titel           VARCHAR(100) NOT NULL,
    instrument      VARCHAR(80) NOT NULL,
    max_teilnehmer  INTEGER NOT NULL CHECK (max_teilnehmer >= 1),
    wochentag       VARCHAR(20) NOT NULL,
    lehrkraft_id    INTEGER NOT NULL REFERENCES lehrkraft(lehrkraft_id)
);

CREATE TABLE teilnahme (
    schueler_id     INTEGER NOT NULL REFERENCES schueler(schueler_id) ON DELETE CASCADE,
    kurs_id         INTEGER NOT NULL REFERENCES kurs(kurs_id) ON DELETE CASCADE,
    seit            DATE NOT NULL,
    status          VARCHAR(20) NOT NULL CHECK (status IN ('aktiv', 'pausiert', 'beendet')),
    PRIMARY KEY (schueler_id, kurs_id)
);
```

```sql
SELECT k.titel, COUNT(t.schueler_id) AS "aktive Teilnehmer" 
  FROM kurs AS k, teilnahme AS t, schueler AS s
  WHERE t.status = 'aktiv' AND s.schueler_id = t.schueler_id AND k.kurs_id = t.kurs_id
  GROUP BY k.titel 
  ORDER BY "aktive Teilnehmer"
```
