## Einführung

### Datenbank SQL Island erstellen

Bevor wir Daten speichern können, müssen wir **Tabellen anlegen**.   Dafür verwenden wir den SQL-Befehl:

```sql
CREATE TABLE tabellenname ( ... );
```

#### Beispiel: Schule

Wir erstellen zwei einfache Tabellen:
- eine Tabelle für Klassen mit den Attributen klassenid und name
- eine Tabelle für Schüler mit den Attributen schuelerid, name, klasse und geschlecht


**Tabelle Klasse**

```sql
CREATE TABLE klasse (
    klassenid INTEGER PRIMARY KEY,
    name TEXT
);
```

**Tablle Schueler**

```sql
CREATE TABLE schueler (
    schuelerid INTEGER PRIMARY KEY,
    name TEXT,
    klasse INTEGER,
    geschlecht CHAR(1)
);
```

##### Erklärung
Innerhalb von `CREATE TABLE (...);`werden die Attribute definiert. Jedes Attribut wird mit Namen und Datentyp festgelegt. Zeilen innerhalb der Definition werden mit Komma getrennt.

| Datentyp | Bedeutung |
| --- | --- |
| INTEGER | ganzzahliger Wert |
| TEXT | Text beliebiger Länge |
| CHAR(1) | Text mit fester Länge (hier Länge 1) |

Ein Attribut, das Primärschlüssel für die Tabelle ist, kann zusätzlich mit `PRIMARY KEY` versehen werden.
